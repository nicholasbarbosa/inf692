```python
from pathlib import Path
import cv2
import time 

DATASET = Path("data/pureNarutoHandSignData")
```

```python
def sleep_time(timer, cap, class_name, count, num_images):
    start = time.time()
    while time.time() - start < timer:
        ret, frame = cap.read()
        segundos_restantes = int(timer - (time.time() - start)) + 1
        cv2.putText(frame, str(segundos_restantes), (50, 100),
                    cv2.FONT_HERSHEY_SIMPLEX, 4, (0, 0, 255), 6)
        cv2.imshow(f"Capturando: {class_name} ({count}/{num_images})", frame)
        cv2.waitKey(1) 
```

```python
def capture_images_from_webcam(class_name, num_images=30, save_dir=DATASET / "train"):
    cap = cv2.VideoCapture(1)
    
    class_dir = save_dir / class_name
    class_dir.mkdir(parents=True, exist_ok=True)
    count = 0

    if not cap.isOpened():
        print("Não foi possível acessar a câmera")
        exit()

    for _ in range(10):
        cap.read()

    while count < num_images:
        ret, frame = cap.read()

        if not ret:
                print("Falha ao capturar imagem da webcam.")
                break

        cv2.imshow(f"Capturando: {class_name} ({count}/{num_images})", frame)
        key = cv2.waitKey(1) & 0xFF

        if key == ord('q'):
            print("Captura interrompida pelo usuário.")
            break

        elif key == ord('c'):
            sleep_time(2, cap, class_name, count, num_images)  # Aguarda 3 segundos antes de capturar a imagem

            ret, frame = cap.read()

            if not ret:
                print("Falha ao capturar imagem da webcam.")
                break

            img_path = class_dir / f"{class_name}_{count:03d}.jpg"
            cv2.imwrite(str(img_path), frame)
            print(f"Imagem salva: {img_path}")

            count += 1

    cap.release()
    cv2.destroyAllWindows()
```

```python
capture_images_from_webcam("bird", num_images=30)
```
