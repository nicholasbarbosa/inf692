```python
from pathlib import Path
import cv2

DATASET = Path("data/pureNarutoHandSignData")
```

```python
def capture_images_from_webcam(class_name, num_images=30, save_dir=DATASET / "train"):
    """
    Captura imagens da webcam para uma classe específica.
    Pressione 'q' para parar a captura.
    """
    cap = cv2.VideoCapture(0)
    count = 0
    class_dir = save_dir / class_name
    class_dir.mkdir(parents=True, exist_ok=True)

    if not cap.isOpened():
        print("Não foi possível acessar a câmera")
        exit()

    for _ in range(10):
        cap.read()

    while count < num_images:
        ret, frame = cap.read()

        if not ret:
            print("Falha ao capturar imagem da webcam.")
            break

        cv2.imshow(f"Capturando: {class_name} ({count}/{num_images})", frame)
        key = cv2.waitKey(1) & 0xFF

        if key == ord('q'):
            print("Captura interrompida pelo usuário.")
            break
        elif key == ord('c'):
            img_path = class_dir / f"{class_name}_{count:03d}.jpg"
            cv2.imwrite(str(img_path), frame)
            print(f"Imagem salva: {img_path}")
            count += 1

    cap.release()
    cv2.destroyAllWindows()
```

```python
capture_images_from_webcam("bird", num_images=30)
```
